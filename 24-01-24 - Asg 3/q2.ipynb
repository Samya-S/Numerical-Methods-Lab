```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0, 0.5, 1, 1.5, 2, 2.5]
y = [0, 0.20, 0.27, 0.30, 0.32, 0.33]

p = 3   # degree of the polynomial
```

```python
n = len(x)

sum_x = sum(x)
sum_y = sum(y)

def sum_xp(xi, k):
    sum = 0
    for i in range(n):
        sum += pow(xi[i],k)
    return sum

def sum_xpy(xi, yi, k):
    sum = 0
    for i in range(n):
        sum += pow(xi[i],k) * yi[i]
    return sum

matx = np.array([[n, sum_x, sum_xp(x, 2), sum_xp(x, 3)], [sum_x, sum_xp(x, 2), sum_xp(x, 3), sum_xp(x, 4)], [sum_xp(x, 2), sum_xp(x, 3), sum_xp(x, 4), sum_xp(x, 5)], [sum_xp(x, 3), sum_xp(x, 4), sum_xp(x, 5), sum_xp(x, 6)]])
m_y = np.array([sum_xpy(x, y, 0), sum_xpy(x, y, 1), sum_xpy(x, y, 2), sum_xpy(x, y, 3)])

matx_inv = np.linalg.inv(matx)

A = np.dot(matx_inv, m_y)

X = np.linspace(0, 2.5, 1000)
Y = A[0] + A[1]*X + A[2]*X*X + A[3]*X*X*X

print(f'y = {A[0]} + {A[1]}x + {A[2]}x^2 + {A[3]}x^3')
```

```python
plt.plot(x, y, 'o')
plt.plot(X, Y)

plt.show()
```
