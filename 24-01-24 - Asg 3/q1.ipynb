```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 1.5, 2, 2.5, 3, 3.5, 4]
y = [25, 31, 27, 28, 36, 35, 32]
```

```python
n = len(x)
sum_x = sum(x)
sum_y = sum(y)
sum_xx = 0
sum_xy = 0
for i in range(n):
    sum_xx += x[i]*x[i]
    sum_xy += x[i]*y[i]
```

```python
b = (n * sum_xy - sum_x * sum_y)/(n * sum_xx - sum_x * sum_x)
print(b)
```

```python
a = (sum_y/(n)) - b * (sum_x/(n))
print(a)
```

```python
# x_pts = np.array(x)
# y_pts = np.array(y)
# plt.plot(x_pts, y_pts, 'o')
plt.plot(x, y, 'o')

X = np.linspace(1, 4, 1000)
Y = a + b * X
plt.plot(X, Y)

plt.show()
```
