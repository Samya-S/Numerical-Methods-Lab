```python
import matplotlib.pyplot as plt 
import numpy as np 
from scipy.interpolate import CubicSpline, interp1d 
plt.rcParams['figure.figsize'] =(12,8) 

x = np.arange(0,1.2,0.2) 
y = x**3 + 2
# natural cubic spline interpolation 
ns = CubicSpline(x, y, bc_type='natural', extrapolate=True) 

# define a new xs 
xs = np.arange(0,1.2,0.2) 
#plot cubic spline and natural cubic spline 
plt.plot(x, y, 'o', label='data') 
plt.plot(xs, (xs)**3 + 2, label='true') 
# plt.plot(xs, cs(xs), label="S") 
plt.plot(xs, ns(xs), label="NS") 
plt.plot(xs, ns(xs,2), label="NS''") 
plt.plot(xs, ns(xs,1), label="NS'") 

plt.legend(loc='upper right', ncol=2) 
plt.title('Cubic/Natural Cubic Spline Interpolation') 
plt.show() 

p1 = 0.5
p2 = 0.7
print("f(0.1) = %s, f(0.3) = %s, f(0.5) = %s" %(ns(0.1,0),ns(0.3,0),ns(0.5,0)))
print("ERROR(0.1) = %s, ERROR(0.3) = %s, ERROR(0.5) = %s" %(abs(ns(0.1,0)-((0.1)**3 + 2)),abs(ns(0.3,0)-((0.3)**3 + 2)),abs(ns(0.5,0)-((0.5)**3 + 2))))
```
